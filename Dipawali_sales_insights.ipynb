# II.a Distribution of Customers

```python
# Creating table with unique customers
filtered_df3 = df3.drop_duplicates(subset='User_ID')
filtered_df3
```

```python
cols=['Gender', 'Marital_Status', 'Zone', 'Orders','Age Group_cat']

@interact
def count_plt(col=list(cols)):
    plt.figure(figsize=(10,5))
    ax=sns.countplot(x=col, data= filtered_df3)
    for bars in ax.containers:
        ax.bar_label(bars)
    plt.ylabel("No_of_customers")   
```

# Insights

|Parametrs|'Gender'| 'Marital_Status'| 'Zone'|'Orders'|'Age Group_cat'|
|--------|--------|-----------------|-------|--------|---------------|
|Max_customers|Female|Unmarried|Central|Two-order|Young_adult(36-45)|

```python
ax= sns.countplot(x='Gender', hue='Marital_Status', data=filtered_df3)
plt.ylabel('No_of_customers')
for bars in ax.containers:
        ax.bar_label(bars)
```

```python
colms=['State', 'Occupation']
@interact
def count_plt(col=list(colms)):
    plt.figure(figsize=(10,7))
    ax=sns.countplot(y=col, data= filtered_df3)
    for bars in ax.containers:
        ax.bar_label(bars)
    plt.xlabel("No_of_customers")
```

- Maximum customers are from UttarPradesh, Maharashtra & Karnataka sates
- Most of the customers are from IT, Heaith care & Aviation professionals respectively.

# II.a Product vs Sales

```python
print('Graphs showing "Average Sales and Avg_No_of_ordres"')
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
ax=sns.barplot(x= 'Product_Category', y='Amount', data=df3)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.grid()
plt.title("Average Sales")

plt.subplot(1,2,2)
ax=sns.barplot(x= 'Product_Category', y='Orders', data=df3)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.grid()
plt.title("Average Orders")

plt.show()
```

```python
amount_df=df3.groupby('Product_Category')['Amount'].sum()/100
orders_df=df3.groupby('Product_Category')['Orders'].sum()
      
print('Graphs showing Sales and Total ordres')

plt.figure(figsize=(15,8))
plt.subplot(1,2,1)
ax= amount_df.plot(kind='bar')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.ylabel('Sales')
for bars in ax.containers:
        ax.bar_label(bars,rotation=90)
plt.title("Total Sales")

plt.subplot(1,2,2)
ax=orders_df.plot(kind='bar')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.ylabel('Total Orders')
for bars in ax.containers:
        ax.bar_label(bars)
plt.title("Total Orders")

plt.show()
```
